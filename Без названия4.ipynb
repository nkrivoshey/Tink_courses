```python
import pandas as pd
import numpy as np 
import string 
import re
import sklearn
from sklearn.model_selection import train_test_split
from sklearn import feature_extraction
from sklearn.pipeline import Pipeline
from sklearn import linear_model
```

```python
df = pd.read_csv('literat.txt',names = ['Текст']).reset_index()
df
```

```python
def del_punct(text):
    for pun in string.punctuation:
        text = text.replace(pun,'').replace('\n','')
    text = text.lower()
    return text
```

```python
df['Текст'] = df['level_1'].apply(del_punct)
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['Текст'],df.index)
```

```python
vector = feature_extraction.text.TfidfVectorizer(ngram_range=(1,1))
```

```python
model = Pipeline([
    ('vector',vector),
    ('linmod',linear_model.LogisticRegression())
])
```

```python
model.fit(X_train,y_train)
pred = model.predict(X_test)
```
